```python
units = int(input("Enter units: "))

bill = 0

if units <= 100:
    bill = units * 1.5
elif units <= 200:
    extra = units - 100
    bill = 100 * 1.5 + extra * 2.5
else:
    extra = units - 200
    bill = 100 * 1.5 + 100 * 2.5 + extra * 4

print("Total bill amount:", bill)
```
