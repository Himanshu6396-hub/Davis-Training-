```python

def simpleInterest(p, r, t):
    si = (p * r * t) / 100
    return si

# Taking input
p = float(input("Enter the principal amount: "))
r = float(input("Enter the rate: "))
t = int(input("Enter the time: "))

# Function call and output
print("Simple interest is:", simpleInterest(p, r, t))
```
