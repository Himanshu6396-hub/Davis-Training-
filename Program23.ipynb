```python
def electricity_bill():
    units = int(input("Enter units consumed: "))

    first = second = third = 0

    if units > 200:
        third = (units - 200) * 4
        units = 200
    if units > 100:
        second = (units - 100) * 2.5
        units = 100
    if units > 0:
        first = units * 1.5

    total = first + second + third
    print("Total bill amount:", total)

# Function call
electricity_bill()
```
