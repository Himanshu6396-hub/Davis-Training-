```python
def simple_interest(p, r, t):
    return (p * r * t) / 100

# Function call with input inside
print("Simple Interest:", simple_interest(
    float(input("Enter P: ")),
    float(input("Enter R: ")),
    float(input("Enter T: "))
))
```
