```python
def si(p, r, t):
    return (p * r * t) / 100

# Input
values = input("Enter P, R, T: ").split()

# Convert and call function
p = float(values[0])
r = float(values[1])
t = float(values[2])

print("Simple Interest:", si(p, r, t))
```
