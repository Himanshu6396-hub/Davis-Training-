```python
def bill(units):
    if units <= 100:
        amount = units * 1.5
    elif units <= 200:
        amount = (100 * 1.5) + (units - 100) * 2.5
    else:
        amount = (100 * 1.5) + (100 * 2.5) + (units - 200) * 4

    # Output
    print("Total bill amount:", amount)


# Input
units = int(input("Enter units consumed: "))

bill(units)
```
